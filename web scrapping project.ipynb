```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
import requests
```

```python
content = requests.get(url).text
```

```python
type(content)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

```python
#content matra garyo vane purai string ma gajyang aauxa
```

```python
tb = soup.find('tbody')
```

```python
tb
```

```python
table_row = tb.find_all('tr')
```

```python
table_row #list ma aai rakhya xa kunai euta lai accesss ni garna sakkiyo list[] liyera
```

```python
table_row[8]
```

```python
all_data = list()
for item in range(8,len(table_row)):
    country_data = list()
    tr = (table_row[item])
    td = tr.find_all("td")
    for dp in td: #scrapes each country data
        country_data.append(dp.text)
        
    all_data.append(country_data) #append each country data to whole data
        
```

```python
all_data[0]
```

```python
# 1:10 14
```

```python
map(lambda x: x[1:10] + [x[14]], all_data)
```

```python
filtered_data = list(map(lambda x: x[1:10] + [x[14]], all_data))
```

```python
filtered_data[0]
```

```python
column_names = ['country',
               'total cases',
               'new cases',
               'total_death',
               'new_death',
               'total recovered',
               'new recovered',
               'active cases',
               'serious',
               'population']
```

```python
import pandas as pd
```

```python
pd.DataFrame(filtered_data)
```

```python
df = pd.DataFrame(filtered_data, columns = column_names)
```

```python
df
```

```python
df.to_csv('covid-jan2.csv',index = False)
```
